```python
days = [1, 2, 3, 4, 5]

sleeping = [7, 8, 7.5, 9, 12]
working = [3, 6, 5, 4, 2]
exercise = [1, 0.5, 1, 0.5, 0.5]
study = [2, 3, 2, 4, 6]

plt.stackplot(days, sleeping, working, exercise, study, labels=['sleeping', 'working', 'exercise', 'study'])
plt.xlabel("X-Axis name")
plt.ylabel("Y-Axis name")
plt.title("My Day TT")
plt.legend()
plt.show()
```

```python
import csv
import matplotlib.pyplot as plt

x = []
y = []

with open('example.txt', 'r') as csvfile:
    plots = csv.reader(csvfile, delimiter=",")
    for row in plots:
        x.append(int(row[0]))
        y.append(int(row[1]))

plt.plot(x, y)
plt.xlabel("X-Axis Name")
plt.ylabel("Y-Axis Name")
plt.title("Loading Data Form FIle")
plt.show()
```
